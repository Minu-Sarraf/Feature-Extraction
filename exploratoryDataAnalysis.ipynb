```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_excel("/ENRSheartfailuredata.xlsx")
data.head()
```

```python
# Checking the missing values
data.isnull().sum()
```

```python
plt.scatter(data.RespiratoryRate ,data.Systolic_BP)
plt.show()
```

```python
#plot the pair plot of salary, balance and age in data dataframe.
sns.pairplot(data = data, vars=['RespiratoryRate','Systolic_BP','HeartRate'])
plt.show()
```

```python
# Creating a matrix using age, salry, balance as rows and columns
data[['RespiratoryRate','Systolic_BP','HeartRate']].corr()

#plot the correlation matrix of salary, balance and age in data dataframe.
sns.heatmap(data[['RespiratoryRate','Systolic_BP','HeartRate']].corr(), annot=True, cmap = 'Reds')
plt.show()
```
